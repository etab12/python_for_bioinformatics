```python
sequence = "ATGC"
if 'ACC' in sequence:
    print("Start codon found")
elif 'TAA' in sequence:
    print("Stop codon found")
else:
    print("No special codons")
```

## الدوال
* مفيدة لإعادة استخدامها مرارا وتكرارا أثناء كتابة الأكواد
*  تختصر التكرار وتجعل الكود أكثر مرونة


```python
def count_seq(sequence):
    count = {}
    for base in sequence:
        if base not in count:
            count[base] = 0
        count[base] += 1
    return count
# استدعاء الدالة
result = count_seq("ATTTCGGC")
print(f"Number of bases in the sequence: {result}")
```

## Pandas


```python
import pandas as pd
```

```python
 
data = {
    'Gene': ['BRCA1', 'TP53', 'EGFR'],
    'Expression': [100, 150, 80],
    'Mutation': ['Yes', 'No', 'Yes']
}
df = pd.DataFrame(data)

# قراءة ملفات موجودة مسبقا 
df_csv = pd.read_csv('Lable_Dataset.csv')
# df_tsv = pd.read_csv('data.tsv', sep='\t')
```

### MatPlot & Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Basic plotting
plt.figure(figsize=(4,2))
plt.plot(df['Gene'], df['Expression'], marker='o')
plt.xlabel('Gene')
plt.ylabel('Expression Level')
plt.title('Gene Expression Levels')
plt.show()
```

```python
# Multiple plots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5)) 
ax1.bar(df['Gene'], df['Expression'])
ax2.scatter(df['Gene'], df['Expression'])
plt.tight_layout()
```

```python
import seaborn as sns

# Statistical visualizations
sns.set_theme(style="ticks", palette="pastel")
sns.boxplot(data=df_csv, x='Position', y='Mutation_Type', hue='Mutation_Type')
plt.xlabel('Position')
plt.ylabel('Mutation')
plt.title('Mutation Types by Position')
plt.show()
```

```python
sns.scatterplot(data=df_csv, x='Position', y='Mutation_Type', hue='Mutation_Type', style='Mutation_Type')
plt.show()
```
